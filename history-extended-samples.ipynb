```python
import os
import pandas as pd
```

# Load dataset

```python
history_country = os.listdir('synthetic-browsing-history')
```

```python
history = pd.DataFrame()

for country in history_country:

    print(country, end=' ')
    
    history_files = os.listdir(os.path.join('synthetic-browsing-history', country))

    for file in history_files:

        file_data = pd.read_csv(os.path.join('synthetic-browsing-history', country, file))

        file_data['history_name'] = file

        file_data['history_country'] = country
    
        history = pd.concat([file_data,history],ignore_index=True,copy=False)

history['history_number'] = history.history_name.apply(lambda x: x.split('_')[1].split('.')[0])

history.history_number = history.history_number.astype('int8')

history.columns
```

# Extended samples

```python
groups = history.groupby(['history_country','history_number'])[['synthetic_time','synthetic_url']]
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
for history, group in groups:
    
    print(f'{history[0]}, history number {history[1]}')
    
    print('Last 5 records')
    
    display(group.tail(5))
```
